```python
# TOP 10 FILMES COM A MAIOR CLASSIFICAÇÃO SEGUNDO O IMDB
df.sort_values(by="IMDB_Rating", ascending=False).head(10)
```

```python
# FILMES COM A MENOR CLASSIFICAÇÃO SEGUNDO O IMDB
df.sort_values(by="IMDB_Rating", ascending=True).head(10)
```

```python
# FILMES COM A MAIOR PONTUAÇÃO [12 FILMES ATINGIRAM A NOTA MÁXIMA]
df.sort_values(by="Meta_score", ascending=False).head(13)
```

```python
# FILMES COM O MAIOR NUMERO DE VOTOS
df.sort_values(by="No_of_Votes", ascending=False).head(10)
```

```python
# FILMES COM O MENOR NUMERO DE VOTOS
df.sort_values(by="No_of_Votes", ascending=False).tail(10)
```

```python
# FILMES SEPARADOS POR GENERO
colors2 = ['#9834eb','red', 'silver', 'orange', 'green', '#eb498c', 'violet', '#014F50', 'gold', 'aquamarine', 'tomato', '#A80874', '#5EF38C', '#3A0328', '#D8F2CF']

df.groupby(by="Genre").size().plot(kind='barh', stacked=True, title='Quantidade de filmes por gênero', color=colors2, figsize=(20, 80))
```

```python
df.groupby(by="Released_Year").size().plot(kind='bar', stacked=True, title='Quantidade de filmes por ano', color=colors2, figsize=(20, 10))
```

```python
# QUANTIDADE DE FILMES POR CERTIFICADOS
df.groupby('Certificate').size().sort_values().plot(kind='barh', stacked=True, title='Certificados', color=colors2, figsize=(13, 7))
```
